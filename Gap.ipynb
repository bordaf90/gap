```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression


```

```python
data = pd.read_csv("gap.tsv", sep='\t')
data.head()
```

### Question 1:

```python
min_data= data[["year","lifeExp"]]  

print(min_data)
```

```python
x=min_data["lifeExp"]
y=min_data["year"]
min_data.plot(x="lifeExp",y="year", style="o")
plt.scatter(x,y,s=100, edgecolor='black',linewidth=0.5,alpha=0.75)
plt.title('lifeExp vs Years \n', color="black",fontsize=15)
plt.xlabel('lifeExp \n',fontsize=15)
plt.ylabel('Year',fontsize=15)


plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = np.array(min_data["lifeExp"]).reshape(-1,1)
y = np.array(min_data["year"]).reshape(-1,1)

x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2)
```

```python
regresion= linear_model.LinearRegression()
regresion.fit(x_train,y_train)
```

```python
print(regresion.score(x_train,y_train)) #El score es bajo, por ende la correlación no es significativa.
```

```python
y_pred= regresion.predict(x_test)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pred,color="red",linewidth=3)
plt.title("Linear Regresion \n",color="black",fontsize=15)
plt.colorbar(orientation="horizontal")

plt.show()

```
