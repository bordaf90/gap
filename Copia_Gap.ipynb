```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
XY_train = np.array(min_data)
z_train= np.array(data["gdpPercap"])
```

```python
# We create a new Linear Regression object.

regr2 = linear_model.LinearRegression()
 
# We train the model, this time, with 2 dimensions.
# We will obtain 2 coefficients, to graph a plane.
regr2.fit(XY_train, z_train)
 
# We make the prediction with which we will have points on the plane found.
z_pred = regr2.predict(XY_train)
 
# The coefficients
print('Coefficients: \n', regr2.coef_)

# Mean square error
print("Mean squared error: %.2f" % mean_squared_error(z_train, z_pred))

# We evaluate the variance score (1.0 being the best possible)
print('Variance score: %.2f' % r2_score(z_train, z_pred))
```

```python

fig = plt.figure()
ax = Axes3D(fig)
 
# We create a mesh, on which we will graph the plane.
xx, yy = np.meshgrid(np.linspace(0, 3500, num=10), np.linspace(10, 60, num=10))
 
# we calculate the values of the plane for the points x and y.
nuevoX = (regr2.coef_[0] * xx)
nuevoY = (regr2.coef_[1] * yy) 
 
# we calculate the corresponding values for z. We must add the interception point.
z = (nuevoX + nuevoY + regr2.intercept_)
 
# We graph the plane
ax.plot_surface(xx, yy, z, alpha=0.5, cmap='hot')
 
ax.scatter(XY_train[:, 0], XY_train[:, 1], z_train, c='black',s=60)
 
ax.scatter(XY_train[:, 0], XY_train[:, 1], z_pred, c='red',s=60)

# "Camera" with which we visualize.

ax.view_init(elev=20., azim=60)
        
ax.set_xlabel('lifeExp \n')
ax.set_ylabel('year \n')
ax.set_zlabel('gdpPercap')
ax.set_title('Regresión Lineal con Múltiples Variables')
```
